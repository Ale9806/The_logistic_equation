## The logistic equation,

The logistic function was introduced in a series of three papers by Pierre François Verhulst between 1838 and 1847, who devised it as a model of population growth by adjusting the exponential growth model, under the guidance of Adolphe Quetelet. Verhulst first devised the function in the mid 1830s, publishing a brief note in 1838, then presented an expanded analysis and named the function in 1844 (published 1845); the third paper adjusted the correction term in his model of Belgian population growth.

$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$
$\frac{dx}{dt} = rx(1-\frac{x}{k}) $ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ 
$ x_{n+1} = rx_n(1-\frac{x_n}{k})$

$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$
Continuous Time Function 
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ 
Discrete Time Function

In order to understand in depth this equation lets solve the Continous time version:

$$\frac{dx}{dt} = rx(1-\frac{x}{k}) $$

$$\frac{dx}{x(1-\frac{x}{k})} = rdt $$

Lets obtain the partial fractions of x:
$$dx(\frac{1}{x(1-\frac{x}{k})})=dx(\frac{A}{x} +\frac{B}{(1-\frac{x}{k})})$$

Lets obtain A $ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; $ Multiply by x
$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ Cancel Terms out
$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ Evaluate in 0

$\frac{1}{x(1-\frac{x}{k})}=\frac{A}{x} +\frac{B}{(1-\frac{x}{k})} \;\;\;\;
\to \;\;\;\; (\frac{x}{x(1-\frac{x}{k})}=\frac{Ax}{x} +\frac{Bx}{(1-\frac{x}{k})})|_{x=0} \;\;\;\;
\to \;\;\;\;(\frac{1}{(1-\frac{x}{k})}=A +\frac{Bx}{(1-\frac{x}{k})})|_{x=0}  \;\;\;\;
\to \;\;\;\;(\frac{1}{(1-\frac{0}{k})}=A +\frac{B0}{(1-\frac{0}{k})}) 
\to 1=A + 0$

Therefore:
$A=1$ 

<br> 

Lets obtain B $ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; $ Multiply by $(1-\frac{x}{k})$
$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ Cancel Terms out
$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ Evaluate in k
<br>

​
$\frac{1}{x(1-\frac{x}{k})}=\frac{A}{x} +\frac{B}{(1-\frac{x}{k})} \;\;\;\;
\to \;\;\;\; (\frac{(1-\frac{x}{k})}{x(1-\frac{x}{k})}=\frac{A(1-\frac{x}{k})}{x} +\frac{B(1-\frac{x}{k})}{(1-\frac{x}{k})})|_{x=k} \;\;\;\;
\to \;\;\;\;(\frac{1}{x}=\frac{A(1-\frac{x}{k})}{x} + B)|_{x=k}  \;\;\;\;
\to \;\;\;\;(\frac{1}{k}= A(0) + B) \;\;\;\;
\to\frac{1}{k}= B$
​

Therefore:
$B=\frac{1}{k}$


Therefore:
$$dx(\frac{1}{x(1-\frac{x}{k})})=dx(\frac{A}{x} +\frac{B}{(1-\frac{x}{k})}) = dx(\frac{1}{x} +\frac{\frac{1}{k}}{(1-\frac{x}{k})})$$

Which means that: 

 $$dx(\frac{1}{x} +\frac{\frac{1}{k}}{(1-\frac{x}{k})}) = rdt $$
 
 Lets obtain the anti derivative in order to solve this:

 $$\int\frac{1}{x}dx +\int \frac{\frac{1}{k}}{(1-\frac{x}{k})}dx = \int rdt $$
 
 Which yields:
 
 $$ ln|x| \;\; - \;\; ln|1-\frac{x}{k}|\;\;  +\;\;  c_1 = rt + c_2 $$
 
 If we make the assumption that $ 0<x(t)<k $
 
  $$ ln(x) \;\; - \;\; ln(1-\frac{x}{k})\;\;  +\;\;  c_1 = rt + c_2 $$
 
By propertie of lograithms we know that $ln(a) - ln(b) = ln(\frac{a}{b}) $ so:

  $$ ln(\frac{x}{1-\frac{x}{k}})\;\;  +\;\;  c_1 = rt + c_2 $$
  
   $$ ln(\frac{x}{1-\frac{x}{k}}) = rt + c_2 - c_1$$
   
   $c_3 =  c_2 - c_1$
   
   $$ ln(\frac{x}{1-\frac{x}{k}}) = rt + c_3$$
   
Lets get rid of $ln$ by elevating everything to the $e$:

 $$ \frac{x}{1-\frac{x}{k}} = e^{rt + c_3}$$
 
 By propertie of exponentials we know that $e^{a + b} =   e^{a}e^{b}$ so:
 
 
 $$ \frac{x}{1-\frac{x}{k}} = e^{c_3}e^{rt}$$
 
 $e^{c_3} =  c_4$
 
  $$ \frac{x}{1-\frac{x}{k}} =  c_4e^{rt}$$
  
Get the inverse

 $$ \frac{1-\frac{x}{k}}{x} =  c_4e^{-rt}  \;\;  \to \;\;  \frac{1}{x}-\frac{x}{xk}=  c_4e^{-rt}   \;\;  \to \;\;  \frac{1}{x}- \frac{1}{k}=  c_4e^{-rt}   $$
 
Add $ \frac{1}{k} $

$$ \frac{1}{x} =  c_4e^{-rt}   + \frac{1}{k} $$

Get the inverse (Again):

$$ x = \frac{1}{ c_4e^{-rt}   + \frac{1}{k}} $$

So the solution for this differential equiton is :

$$ x(t)= \frac{1}{ c_4e^{-rt}   + \frac{1}{k}}  \;\;\;\;   0<x(t)<k  $$

The most famous  representation for this function  is the sigmoid function in nerual networks, in which $k=1$ and c_4 =1:

$$ x(t)= \frac{1}{e^{-rt}   + 1}  \;\;\;\;   0<x(t)<1  $$

This function could come from the following equation:

$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$
$\frac{dx}{dt} = rx(1-x) $ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ 
$ x_{n+1} = rx_n(1-x_n)$

$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$
Continuous Time Function 
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ 
Discrete Time Function

```python
def logistic_function(xn,r,time):
    """ Discrete Logistic function"""
    xn_1 = []       # Init xn_1 vector (the derivative of xn)
    xn_1.append(xn) # append initial condition
    for t in range(1,len(time)):
        a = r*xn*(1-xn)
        xn = a
        xn_1.append(xn)
    return np.array(xn_1)

n=20                           # Right boundary
time = np.linspace(0,n,n+1)    # Create vector
xn = 0.4                       # Set initial xn, keep in mind that x: 0 < x < 1    
r = 2.6                        # Set rate 
xn_1 = logistic_function(xn,r,time)  # Obtain function

### PLOT ###
plt.figure(figsize=(10,6))
plt.title(" Discrete Logistic function (Derivative of xn)")
plt.plot(time,xn_1)
plt.xlabel("Time")
plt.ylabel(" $x_{n+1}$ ")
plt.show()
```

## Lets analyze this in depth: 
Lets begin by analyzing what what happens if the initial condition changes, keep in mind that:  $ 0<x(t)<1  $

```python
plt.figure(figsize=(14,6))
plt.title("Behavior of rate $r$")

n=6                          # Right boundary
time = np.linspace(0,n,n+n+2) # Create vector
xn = 0.9                      # Set initial xn, keep in mind that x: 0 < x < 1    
rest = 0.07                   # Constant to rest                   
number_of_plots=9             # Number of plots to plot
r = 2.6                       # Intial r

###### Create plots ######
for i in range(0,number_of_plots):
    xn  = xn  - rest                                  # Substract to r
    xn_1 = logistic_function(xn,r,time)           # Calcualte new line
    if i != number_of_plots-1:                    # Plot everything except the last plot
        plt.plot(time,xn_1,label=f"xn_o = {round(xn,2)}") # plot it
    else:                                         # Change linesytle for last plot 
        plt.plot(time,xn_1,linestyle="--",label=f"xn_o = {round(xn,2)}")
        

###### Make Plots Pretty ######
plt.legend(loc="upper right")
plt.xlabel("Time in years")
plt.ylabel(" $x_{n+1}$ ")
plt.show()
```

Notice that no matter what the inital condition is, the function will always reach the the same steady state (about 6.5)

```python
#### We have seen that this function is only stable from -1 to 3, lets  analyze this points  ####
n = 100
time = np.linspace(0,n,n*100 +1)  # Create vector
xn = 0.4                       # Set initial xn, keep in mind that x: 0 < x < 1    
r =  -1                        # Intial r
increment = 0.25               # Factor to substract to r each iteration
r_vec = []                     # Store r's in a vector
###### Create plots ######
points = []                    # Create an empty list to store points
for i in range(0,17):          # iterate from -1 to 3
    xn_1 = logistic_function(xn,r,time)  # Calcualte new line
    points.append(xn_1[-1])              # Append limit of function to points
    r_vec.append(r)                      # Append r to r vector
    r = r  + increment                   # Increment
    
## PLOT ###
plt.figure(figsize=(10,6))               # Create a figure 
plt.title(" $r \;\; $ vs $  \;\; \lim_{n -> \infty} x(n)$  ")
plt.plot(r_vec,points)                   # Plot values


# r is 3 here 
### Lets analyze the behavior beyond 3 ###
r=3                                       # Set r to 3 
for i in range(0,2500):                    # Iterate 500 times
    xn_1 = logistic_function(xn,r,time)   # Calcualte function
    fractal_points = []                   # Create an empty list to save points   (since we know it does not reach steady state)
    for j in range(-10,0,1):              # See all the values from n-10 to n this values oscilate
        if xn_1[j] in set(fractal_points): # Iterate ovr them and see if we have seen them before
            pass                           # If vale is not new, ignore
        else:                              # Else, append it to list of fractal points
            fractal_points.append(xn_1[j])
    for pf in fractal_points:             # For every differnt fractal point
        plt.plot(r,pf,".")                # Plot it in point r
    r = r  + 0.005                         # Increment 

# Do the same but for r backwards for -1
r = -1                                    # Set r to -1
for i in range(0,2500):                    # Iterate 500 times
    xn_1 = logistic_function(xn,r,time)   # Calcualte function
    fractal_points = []                   # Create an empty list to save points   (since we know it does not reach steady state
    for j in range(-10,0,1):              # See all the values from n-10 to n this values oscilate
        if xn_1[j] in set(fractal_points): # Iterate ovr them and see if we have seen them before
            pass                           # If vale is not new, ignore
        else:                              # Else, append it to list of fractal points
            fractal_points.append(xn_1[j]) 
    for pf in fractal_points:              # For every differnt fractal point
        plt.plot(r,pf,".")                 # Plot it in point r
    r = r  - 0.005                           # Decrement 
    
### MAKE PLOT PRETTY ##
plt.ylabel("$ \lim_{n -> \infty} x(n)$")
plt.xlabel("$r$")
plt.show()
```

This is no oridnary fractal, this is actually a subset of the Mandelbrot set times -1. where the blue line represents the values of the main cardoid of the maandelbrot set times -1.

This subset is nown as the birfurcation diagram. Lets dig deeper 

```python
#### We have seen that this function is only stable from -1 to 3, lets  analyze this points  ####
n = 100
time = np.linspace(0,n,n*100 +1)  # Create vector
xn = 0.4                       # Set initial xn, keep in mind that x: 0 < x < 1    
r =  -1                        # Intial r
increment = 0.25               # Factor to substract to r each iteration
r_vec = []                     # Store r's in a vector
###### Create plots ######
points = []                    # Create an empty list to store points
for i in range(0,17):          # iterate from -1 to 3
    xn_1 = logistic_function(xn,r,time)  # Calcualte new line
    points.append(xn_1[-1])              # Append limit of function to points
    r_vec.append(r)                      # Append r to r vector
    r = r  + increment                   # Increment
    
## PLOT ###
plt.figure(figsize=(10,6))               # Create a figure 
plt.title(" $r \;\; $ vs $  \;\; \lim_{n -> \infty} x(n)$  ")
plt.plot(r_vec,points)                   # Plot values


# r is 3 here 
### Lets analyze the behavior beyond 3 ###
r=3                                       # Set r to 3 
for i in range(0,2500):                    # Iterate 500 times
    xn_1 = logistic_function(xn,r,time)   # Calcualte function
    fractal_points = []                   # Create an empty list to save points   (since we know it does not reach steady state)
    for j in range(-10,0,1):              # See all the values from n-10 to n this values oscilate
        if xn_1[j] in set(fractal_points): # Iterate ovr them and see if we have seen them before
            pass                           # If vale is not new, ignore
        else:                              # Else, append it to list of fractal points
            fractal_points.append(xn_1[j])
    for pf in fractal_points:             # For every differnt fractal point
        plt.plot(r,pf,".")                # Plot it in point r
    r = r  + 0.005                         # Increment 

# Do the same but for r backwards for -1
r = -1                                    # Set r to -1
for i in range(0,2500):                    # Iterate 500 times
    xn_1 = logistic_function(xn,r,time)   # Calcualte function
    fractal_points = []                   # Create an empty list to save points   (since we know it does not reach steady state
    for j in range(-10,0,1):              # See all the values from n-10 to n this values oscilate
        if xn_1[j] in set(fractal_points): # Iterate ovr them and see if we have seen them before
            pass                           # If vale is not new, ignore
        else:                              # Else, append it to list of fractal points
            fractal_points.append(xn_1[j]) 
    for pf in fractal_points:              # For every differnt fractal point
        plt.plot(r,pf,".")                 # Plot it in point r
    r = r  - 0.005                           # Decrement 
    
### MAKE PLOT PRETTY ##
plt.axvline(x=3.2,linestyle="--",color="k")
plt.axhline(y=0.51,linestyle="--",color="k")
plt.axhline(y=0.8,linestyle="--",color="k")
plt.ylabel("$ \lim_{n -> \infty} x(n)$")
plt.xlabel("$r$")
plt.show()
```
